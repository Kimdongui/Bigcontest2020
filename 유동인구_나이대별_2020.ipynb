```python
# day_of_month_1902 = 28
# day_of_month_1903 = 31
# day_of_month_1904 = 30
# day_of_month_1905 = 31
# day_of_month_2002 = 29
# day_of_month_2003 = 31
# day_of_month_2004 = 30
# day_of_month_2005 = 31
# num_hdong = 69
# sum_index = day_of_month * num_hdong

# 특정 년/월/동 데이터프레임 만들기

def Make_Age_HDONG(month, num_HD):
    
    HDONG = []
    
    if month == 1902:
        day = 28
        base_df = Flow_Age_1902
    elif month == 1903:
        day = 31
        base_df = Flow_Age_1903
    elif month == 1904:
        day = 30
        base_df = Flow_Age_1904
    elif month == 1905:
        day = 31
        base_df = Flow_Age_1905
    elif month == 2002:
        day = 29
        base_df = Flow_Age_2002
    elif month == 2003:
        day = 31
        base_df = Flow_Age_2003
    elif month == 2004:
        day = 30
        base_df = Flow_Age_2004
    elif month == 2005:
        day = 31
        base_df = Flow_Age_2005
    else:
        print("1902~1905, 2002~2005")
    num_hdong = 69
    sum_index = day * num_hdong
    
    for index in range(0, sum_index, day):
        HDONG.append(base_df[index:index+day])
        
    return HDONG[num_HD]
```

```python
# 특정 동의 나이대 별 유동인구 수
# x 축 = 일(며칠)
# y 축 = 유동인구 수
# 행정동 기준으로 나눠준 데이터프레임을 인자로 넣어야함.

def Make_Age_Plot_HDONG(HD):
    fig, axes = plt.subplots(nrows=5,ncols=4)
    fig.set_size_inches(12, 18)
    age = 0
    sex_age = 0
    x = 0
    y = 0
    x_chuc = HD[HD.columns[1]] % HD[HD.columns[0]]
    y_limit = [5000 for i in range(len(x_chuc))]
    mpl.rc('font', family='NanumGothic')
    
    while sex_age <= 14:
        for x in range(4):
            for y in range(4):
                if x == 3 & y == 3:
                    continue
                axes[x][y].plot(x_chuc, HD[HD.columns[sex_age+4]])
                axes[x][y].plot(x_chuc, HD[HD.columns[sex_age+19]])
                axes[x][y].plot(x_chuc, y_limit)
                if age < 70:
                    axes[x][y].set_title("{}~{}세".format(age,age+4))
                else:
                    axes[x][y].set_title("70세 Up")
                age += 5
                sex_age += 1
    
    axes[4][0].plot(x_chuc, HD[HD.columns[34]])
    axes[4][0].plot(x_chuc, HD[HD.columns[35]])
    axes[4][0].plot(x_chuc, y_limit)
    axes[4][0].set_title("연령대 합")
    axes[4][1].plot(x_chuc, HD[HD.columns[36]], c='black')
    axes[4][1].set_title("연령대+성 합")

    plt.xlim(0, 30)
    plt.ylim(0, 1000000)
    plt.suptitle(str(HD.STD_YM.values[0]) + ' ' + str(HD.HDONG_NM.values[0])
                 + '. < 하늘:남자, 주황:여자 >', fontsize=20)
    plt.show()
```

```python
# 전체 나이대 합 기준, 행정동 별 유동인구 수
# x 축 = 일(며칠)
# y 축 = 유동인구 수
# 행정동 기준으로 나누기 전!! 기존 데이터 그대로를 인자로 넣어야 함!!

def Make_SumAge_Plot_HDONG(FA):
    fig, axes = plt.subplots(nrows=10,ncols=7)
    fig.set_size_inches(12, 30)
    idx_HD = 0
    x = 0
    y = 0
    HDONG = []
    
    if FA.STD_YM.values[0] == 201902:
        day = 28
    elif FA.STD_YM.values[0] == 201903:
        day = 31
    elif FA.STD_YM.values[0] == 201904:
        day = 30
    elif FA.STD_YM.values[0] == 201905:
        day = 31
    elif FA.STD_YM.values[0] == 202002:
        day = 29
    elif FA.STD_YM.values[0] == 202003:
        day = 31
    elif FA.STD_YM.values[0] == 202004:
        day = 30
    elif FA.STD_YM.values[0] == 202005:
        day = 31
    else:
        print("201902~201905, 202002~202005")

    num_hdong = 69
    sum_index = day * num_hdong
    
    for index in range(0, sum_index, day):
        HDONG.append(FA[index:index+day])
        
    x_chuc = HDONG[idx_HD][HDONG[idx_HD].columns[1]] % HDONG[idx_HD][HDONG[idx_HD].columns[0]]
    y_limit = [300000 for i in range(len(x_chuc))]
    
    while idx_HD < num_hdong:
        for x in range(10):
            for y in range(7):
                if idx_HD >= num_hdong:
                    break
                axes[x][y].plot(x_chuc, HDONG[idx_HD][HDONG[idx_HD].columns[34]])
                axes[x][y].plot(x_chuc, HDONG[idx_HD][HDONG[idx_HD].columns[35]])
                axes[x][y].plot(x_chuc, HDONG[idx_HD][HDONG[idx_HD].columns[36]])
                axes[x][y].plot(x_chuc, y_limit, color='black')
                axes[x][y].set_title(HDONG[idx_HD].HDONG_NM.values[0])
                idx_HD += 1

    plt.xlim(0, 30)
    plt.ylim(0, 10000000)
    plt.suptitle(str(FA.STD_YM.values[0]), fontsize=20)
    mpl.rc('font', family='NanumGothic')
    plt.show()

# list index out of range 라고 뜨면서 된다. 이유는 몰겠.
```

```python
def Make_Age_SumHDONG(FA):
    base_df = FA
    Sum_HDONG = base_df.groupby(['STD_YMD'])['MAN_FLOW_POP_CNT_0004'].sum()
    Sum_HDONG = pd.DataFrame(Sum_HDONG)
    base_df_columns = base_df.columns[4:37].tolist()
    
    for i in range(32):
        temp = base_df.groupby(['STD_YMD'])[base_df_columns[i+1]].sum()
        temp = pd.DataFrame(temp)
        Sum_HDONG[base_df.columns[i+5]] = temp[temp.columns[0]]
    return Sum_HDONG

Make_Age_SumHDONG(Flow_Age_2003)
```

```python
def Make_Age_Plot_SumHDONG(Sum_HDONG):
    fig, axes = plt.subplots(nrows=5,ncols=4)
    fig.set_size_inches(12, 18)
    age = 0
    sex_age = 0
    x = 0
    y = 0
    y_limit = [300000 for i in range(len(Sum_HDONG.index))]
    
    while sex_age <= 14:
        for x in range(4):
            for y in range(4):
                if x == 3 & y == 3:
                    continue
                axes[x][y].plot(Sum_HDONG.index, Sum_HDONG[Sum_HDONG.columns[sex_age]])
                axes[x][y].plot(Sum_HDONG.index, Sum_HDONG[Sum_HDONG.columns[sex_age+15]])
                axes[x][y].plot(Sum_HDONG.index, y_limit)
                if age < 70:
                    axes[x][y].set_title("{}~{}세".format(age,age+4))
                else:
                    axes[x][y].set_title("70세 Up")
                age += 5
                sex_age += 1
    
    axes[4][0].plot(Sum_HDONG.index, Sum_HDONG[Sum_HDONG.columns[30]])
    axes[4][0].plot(Sum_HDONG.index, Sum_HDONG[Sum_HDONG.columns[31]])
    axes[4][0].set_title("나이대 합")
    axes[4][1].plot(Sum_HDONG.index, Sum_HDONG[Sum_HDONG.columns[32]], c='black')
    axes[4][1].set_title("나이대+성 합")
    
    plt.xlim(0, 30)
    plt.ylim(0, 1000000)
    mpl.rc('font', family='NanumGothic')
    plt.suptitle(Sum_HDONG.index.tolist().pop(0) // 100, fontsize=20)
    plt.show()
```

```python
# 그 달의 연령대 총 유동인구

def Make_Age_Sum(Sum_HDONG):
    plt.rc('font', family='NanumGothic')
    x_Mann = [0,5,10,15,20,25,30,35,40,45,50,55,60,65,70]
    x_Man = np.array(x_Mann)
    y_Man = []
    for i in range(len(x_Man)):
        y_Man.append(Sum_HDONG[Sum_HDONG.columns[i]].sum())

    y_WMan = []
    for i in range(len(x_Man)):
        y_WMan.append(Sum_HDONG[Sum_HDONG.columns[i+15]].sum())
    
    y_limit = [10000000 for i in range(len(x_Man))]

    plt.figure(figsize=(18,8))
    plt.bar(x_Man+0, y_Man, color='skyblue', width=1.5, label='Man')
    plt.bar(x_Man+1, y_WMan, color='orange', width=1.5, label='Woman')
    plt.plot(x_Man, y_limit, label='10,000,000')
    plt.suptitle(str(Sum_HDONG.index[0]//100) + '월의 총 유동인구 수 <연령대별>', fontsize=20)
    plt.legend()
    plt.show()
```
