```python
#  합친 데이터를 서울시 중구/노원구로 자름.
#  서울시 중구:15동 . 노원구:19동 . 

day = 121 # 2020.2~5월 총 일수


population_2020 = population_2020.sort_values(by=['HDONG_CD','STD_YMD'])
Seoul_Joongu = population_2020[:day*15]
Seoul_Joongu = Seoul_Joongu.reset_index(drop=True)
Seoul_Nowongu = population_2020[day*15:day*34]
Seoul_Nowongu = Seoul_Nowongu.reset_index(drop=True)

Seoul_Joongu
```

```python
temp = Seoul_Joongu.groupby(Seoul_Joongu.columns[1])[Seoul_Joongu.columns[4:34]].sum()
temp = pd.DataFrame(temp)
temp = temp.reset_index()
temp
```

```python
S_J_sum = pd.DataFrame(temp['STD_YMD'])
S_J_sum['남20대'] = temp[temp.columns[1:7]].sum(axis=1)

age = 30
for i in range(0,8,2):
    S_J_sum['남{}대'.format(age)] = temp[temp.columns[i+7:i+9]].sum(axis=1)
    age += 10
S_J_sum['남70이상'] = temp[temp.columns[15]]
    
S_J_sum['여20대'] = temp[temp.columns[16:22]].sum(axis=1)

age = 30
for i in range(0,8,2):
    S_J_sum['여{}대'.format(age)] = temp[temp.columns[i+22:i+24]].sum(axis=1)
    age += 10
S_J_sum['여70이상'] = temp[temp.columns[30]]
    
S_J_sum
```

-------------------------------< 카드매출 데이터 >---------------------------------

```python
text_resident = pd.read_csv("./CARD_SPENDING_RESIDENT.txt", sep = "\t")
```

```python
text_resident.columns = ['기준일자','구코드','행정동코드','업종코드','성별코드','나이코드','이용건수','이용금액']
text_resident
```

```python
b = {140: '서울 중구', 350: '서울 노원구', 260: '대구 수성구', 110: '대구 중구'}

temp2 = copy.copy(text_resident)
temp2["구코드"].replace(b,inplace=True)
temp2 = pd.DataFrame(temp2.groupby(["구코드","기준일자"])["이용금액"].sum())
temp2_S_J = temp2.loc['서울 중구'][120:241]
temp2_S_J = temp2_S_J.reset_index()
temp2_S_J
```

--------------< 유동인구 + 카드매출 >-----------------

```python
S_J_sum['총이용금액'] = temp2_S_J['이용금액']
S_J_sum
```

```python
x_data = S_J_sum[S_J_sum.columns[1:13]]
y_data = S_J_sum['총이용금액']
```

```python
print(x_data.shape, y_data.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn import linear_model

sk_lr = linear_model.LinearRegression(normalize=False)
sk_lr.fit(X_train, y_train)

sk_lr.intercept_ , sk_lr.coef_
```

```python
print("선형회귀 훈련 세트 점수: {}".format(sk_lr.score(X_train, y_train)))
print("선형회귀 테스트 세트 점수: {}".format(sk_lr.score(X_test, y_test)))
```

```python
from sklearn.linear_model import Lasso, Ridge
```

```python
ridge = Ridge(fit_intercept=True, alpha=1)
ridge.fit(X_train,y_train)
#lasso = Lasso(fit_intercept=True, alpha=0.5)
```

```python
ridge.intercept_ , ridge.coef_
```

```python
print("릿지 훈련 세트 점수: {}".format(ridge.score(X_train, y_train)))
print("릿지 테스트 세트 점수: {}".format(ridge.score(X_test, y_test)))
```

```python
lasso = Lasso(fit_intercept=True, alpha=1)
lasso.fit(X_train,y_train)
```

```python
print("랏소 훈련 세트 점수: {}".format(lasso.score(X_train, y_train)))
print("랏소 테스트 세트 점수: {}".format(lasso.score(X_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier(n_estimators=10,random_state=10).fit(X_train,y_train)
print("RandomForest")
print("훈련 세트 정확도: {}".format(RF.score(X_train,y_train)))
print("테스트 세트 정확도: {}".format(RF.score(X_test,y_test)))
```
