```python
SD_daily = pd.DataFrame(regional_daily.loc[0]).T

for i in range(1, 3492):
    if regional_daily['region'].values[i] == '서울' or regional_daily['region'].values[i] == '대구':
        SD_daily = SD_daily.append(regional_daily.loc[i])
    else:
        continue
        
SD_daily = SD_daily.sort_values(by=['region','date'])
SD_daily = SD_daily.reset_index(drop=True)
D_daily = SD_daily[0:194]
S_daily = SD_daily[194:].reset_index(drop=True)
D_daily
```

```python
S_daily
```

```python
fig = plt.figure(figsize=(18,18))

plt.plot(D_daily.index, D_daily[D_daily.columns[2]], label='확진자 수')
plt.plot(D_daily.index, D_daily[D_daily.columns[3]], label='사망자 수')
plt.plot(D_daily.index, D_daily[D_daily.columns[4]], label='완치자 수')
plt.title('대구 : 20.02.17 ~ 20.08.28', fontsize=24)
plt.legend(fontsize=20)
plt.show()
```

```python
fig = plt.figure(figsize=(18,18))

plt.plot(S_daily.index, S_daily[S_daily.columns[2]], label='확진자 수')
plt.plot(S_daily.index, S_daily[S_daily.columns[3]], label='사망자 수')
plt.plot(S_daily.index, S_daily[S_daily.columns[4]], label='완치자 수')
plt.title('서울 : 20.02.17 ~ 20.08.28', fontsize=24)
plt.legend(fontsize=20)
plt.show()
```
