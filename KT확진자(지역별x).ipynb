```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import matplotlib.font_manager as fm
mpl.rcParams["axes.unicode_minus"] = False
mpl.rc("font", family = "Gulim")
```

```python
kt = pd.read_csv("../Desktop/새 폴더/2020빅콘테스트 문제데이터(혁신아이디어분야)/kr_daily.csv")
```

```python
fig = plt.figure(figsize=(13, 25))

fig.add_subplot(6,2,1)
plt.plot(kt["confirmed"], label = "누적 확진자")
plt.plot(kt["released"], label = "누적 완치자")
plt.grid(b=None, which='major', axis="both")
plt.legend()

fig.add_subplot(6,2,2)
plt.plot((kt["confirmed"] - kt["released"]), label = "현재 확진자", color = "red")
plt.grid(b=None, which='major', axis="both")
plt.legend()

fig.add_subplot(6,2,3)
plt.plot(kt["tested"], label = "테스트 수")
plt.plot(kt["negative"], label = "음성 반응")
plt.grid(b=None, which='major', axis="both")
plt.legend()

fig.add_subplot(6,2,4)
plt.plot((kt["tested"] - kt["negative"]), label = "누적 양성 반응", color = "red")
plt.grid(b=None, which='major', axis="both")
plt.legend()

fig.add_subplot(6,2,5)
plt.plot(kt["death"], label = "누적 사망자 수")
plt.grid(b=None, which='major', axis="both")
plt.legend()

plt.show()
```

```python
date_list = list(kt["date"])
temp_list = []

for i in range(222):
    if i == 0:
        temp = int(kt["death"].loc[kt["date"] == date_list[i]])
    else:
        temp = int(kt["death"].loc[kt["date"] == date_list[i]]) - int(kt["death"].loc[kt["date"] == date_list[i - 1]])
    temp_list.append(temp)
    
daily_death = pd.DataFrame(temp_list)

fig = plt.figure(figsize = (25, 13))
plt.title("일일 사망자 수", fontsize = "24")
plt.plot(daily_death, label = "사망자 수")
plt.legend(fontsize = "24")
plt.show()
```
